## For creating isArrythmia target column(merging both dataset)

```python
import pandas as pd
import ast

# ---------------- Step 1: Load datasets ----------------
df = pd.read_csv("ptbxl_database.csv")
scp = pd.read_csv("scp_statements.csv")

# ---------------- Step 2: Create mapping dict ----------------
# Map SCP code → diagnostic_class
scp_dict = scp[scp["diagnostic_class"].notnull()].set_index("Unnamed: 0")["diagnostic_class"].to_dict()

# ---------------- Step 3: Parse scp_codes ----------------
df["scp_codes"] = df["scp_codes"].apply(ast.literal_eval)

# ---------------- Step 4: Map to diagnostic classes ----------------
# Convert scp_codes dict → list of diagnostic classes
def get_classes(code_dict):
    return list({scp_dict.get(code) for code in code_dict.keys()} - {None})

df["target_classes"] = df["scp_codes"].apply(get_classes)

# ---------------- Step 5: Create binary target (Arrhythmia vs Normal) ----------------
arrhythmia_classes = ["MI", "STTC", "CD", "HYP"]

def is_arrhythmia(classes):
    if not classes:
        return None
    if "NORM" in classes and len(classes) == 1:
        return 0
    elif any(cls in arrhythmia_classes for cls in classes):
        return 1
    else:
        return None

df["isArrhythmia"] = df["target_classes"].apply(is_arrhythmia)

# ---------------- Step 6: Select final useful columns ----------------
final_df = df[[
    "ecg_id", "patient_id", "age", "sex",
    "filename_lr", "filename_hr",   # ECG signal files
    "target_classes", "isArrhythmia"
]]

print("Final dataset shape:", final_df.shape)
print(final_df.head())
```

#### Define the load_ecg function

```python
import wfdb
import os
# Root folder where ptbxl_database.csv, records100/, records500/ exist
DATA_PATH = "C:/Users/hp/Desktop/PROD"
def load_ecg(index, resolution="lr"):
    """
    Load an ECG signal + metadata + label from PTB-XL by row index in final_df.
    
    Parameters:
        index (int): Row index in final_df
        resolution (str): "lr" (100 Hz) or "hr" (500 Hz)
        
    Returns:
        signals (ndarray): ECG signal (timepoints × leads)
        fields (dict): Metadata from WFDB (fs, sig_name, etc.)
        label (int): 1 = Arrhythmia, 0 = Not Arrhythmia
    """   # pick correct resolution
    if resolution == "lr":
        rel_path = final_df.loc[index, "filename_lr"]
    elif resolution == "hr":
        rel_path = final_df.loc[index, "filename_hr"]
    else:
        raise ValueError("resolution must be 'lr' or 'hr'")
   # build full path
    record = os.path.join(DATA_PATH, rel_path)
    
    # load ECG signal
    signals, fields = wfdb.rdsamp(record)
    
    # fetch label
    label = final_df.loc[index, "isArrhythmia"]
    
    return signals, fields, label
```

###  -->target_classes vs isArrhythmia

####  isArrhythmia already converts this into binary labels (0,1).

```python
#Drop invalid rows
#Drop rows where isArrhythmia is NaN

# Drop rows with missing labels
final_df = final_df.dropna(subset=["isArrhythmia"])
print("Shape after dropping invalid labels:", final_df.shape)
```

```python
#Fix age column
#Most patients are between 17–90 years.
#Ages > 120 are unrealistic → set them as NaN and impute with median

# Replace unrealistic ages (e.g., > 120) with NaN
final_df.loc[final_df["age"] > 120, "age"] = None

# Impute missing/invalid ages with median
final_df["age"] = final_df["age"].fillna(final_df["age"].median())
```

```python
#Patient-level split
#Ensure same patient ECGs stay in the same dataset.
#We’ll split by patient_id not by row.

from sklearn.model_selection import train_test_split

# Get unique patients
unique_patients = final_df["patient_id"].unique()

# Split patients into train/test
train_patients, test_patients = train_test_split(
    unique_patients, test_size=0.2, random_state=42
)

# Assign rows
train_meta = final_df[final_df["patient_id"].isin(train_patients)]
test_meta  = final_df[final_df["patient_id"].isin(test_patients)]

print("Train shape:", train_meta.shape)
print("Test shape:", test_meta.shape)
```

```python
import neurokit2 as nk

def plot_beat_overlay(index, lead="II", window=200):
    """
    Detect R-peaks and overlay beats for one ECG lead.
    
    Parameters:
    - index: row index from final_df
    - lead: which lead to analyze (default: "II")
    - window: samples before and after R-peak (default: 200 → 2s at 100Hz)
    """
    # Load ECG
    rel_path = final_df.loc[index, "filename_lr"]
    record = os.path.join("C:/Users/hp/Desktop/PROD", rel_path)
    signals, fields = wfdb.rdsamp(record)
    
    # Find lead index
    if lead not in fields["sig_name"]:
        print(f"Lead {lead} not found in record")
        return
    lead_idx = fields["sig_name"].index(lead)
    ecg_signal = signals[:, lead_idx]
    
    # Detect R-peaks using NeuroKit2
    _, rpeaks = nk.ecg_peaks(ecg_signal, sampling_rate=100)
    r_locs = rpeaks["ECG_R_Peaks"]
    
    # Extract beats around each R-peak
    beats = []
    for r in r_locs:
        start = max(0, r - window//2)
        end = min(len(ecg_signal), r + window//2)
        segment = ecg_signal[start:end]
        if len(segment) == window:  # keep consistent length
            beats.append(segment)
    
    # Overlay all beats
    plt.figure(figsize=(10, 6))
    for beat in beats:
        plt.plot(beat, color="blue", alpha=0.3)
    plt.title(f"Overlayed Beats - Lead {lead} | ECG ID {final_df.loc[index,'ecg_id']}")
    plt.xlabel("Samples (100Hz)")
    plt.ylabel("Amplitude (mV)")
    plt.show()

# Example: Normal
plot_beat_overlay(norm_idx, lead="II")

# Example: Arrhythmia
plot_beat_overlay(arr_idx, lead="II")

#✅ What this does:
#Detects R-peaks.
#Cuts out fixed-length windows around each heartbeat.
#Overlays them to check if beats are consistent (normal) or irregular (arrhythmia).
#👉 In normal ECG, you’ll see beats stack nicely.
# In arrhythmia, they’ll look misaligned, irregular, or distorted.
```

```python
#The idea is:
#->Detect R-peaks.
#->Segment beats around each R-peak.
#->Stack beats → compute mean + standard deviation.
#Plot average waveform with shaded area showing variability.

#Here’s the code for Lead II

import numpy as np

def plot_avg_beat(index, lead="II", window=200):
    """
    Compute and plot average beat morphology for a given ECG lead.
    
    Parameters:
    - index: row index from final_df
    - lead: lead name (default: "II")
    - window: number of samples per beat segment (default: 200 → 2s at 100Hz)
    """
    # Load ECG
    rel_path = final_df.loc[index, "filename_lr"]
    record = os.path.join("C:/Users/hp/Desktop/PROD", rel_path)
    signals, fields = wfdb.rdsamp(record)

    # Find lead index
    if lead not in fields["sig_name"]:
        print(f"Lead {lead} not found in record")
        return
    lead_idx = fields["sig_name"].index(lead)
    ecg_signal = signals[:, lead_idx]

    # Detect R-peaks
    _, rpeaks = nk.ecg_peaks(ecg_signal, sampling_rate=100)
    r_locs = rpeaks["ECG_R_Peaks"]

    # Extract beats
    beats = []
    for r in r_locs:
        start = max(0, r - window//2)
        end = min(len(ecg_signal), r + window//2)
        segment = ecg_signal[start:end]
        if len(segment) == window:
            beats.append(segment)
    
    if len(beats) == 0:
        print("No beats detected.")
        return
    
    beats = np.array(beats)
    mean_beat = np.mean(beats, axis=0)
    std_beat = np.std(beats, axis=0)

    # Plot average + variability
    plt.figure(figsize=(10, 6))
    plt.plot(mean_beat, color="red", label="Average Beat")
    plt.fill_between(range(window), mean_beat-std_beat, mean_beat+std_beat, 
                     color="red", alpha=0.2, label="±1 Std Dev")
    plt.title(f"Average Beat Morphology - Lead {lead} | ECG ID {final_df.loc[index,'ecg_id']}")
    plt.xlabel("Samples (100Hz)")
    plt.ylabel("Amplitude (mV)")
    plt.legend()
    plt.show()

# Example: Normal
plot_avg_beat(norm_idx, lead="II")

# Example: Arrhythmia
plot_avg_beat(arr_idx, lead="II")

#Normal ECG → average waveform is smooth, QRS complex sharp, low variance.
#Arrhythmia ECG → average waveform distorted, higher variance (shaded area wider).
```

```python
import numpy as np
import wfdb
from scipy.signal import butter, filtfilt
```

```python
#Define preprocessing functions
#Baseline wander removal – removes slow drift caused by respiration or movement.
#Bandpass filter – keeps frequencies important for ECG (0.5–40 Hz) and removes noise.
#Per-lead normalization – standardizes each lead to mean 0, std 1.

# Bandpass filter function
def bandpass_filter(signal, fs=100, lowcut=0.5, highcut=40):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(2, [low, high], btype='band')
    return filtfilt(b, a, signal)

# Preprocess one ECG signal (all 12 leads)
def preprocess_ecg(signals, fs=100):
    """
    Input:
        signals: ndarray, shape (timepoints, 12)
    Output:
        preprocessed_signals: ndarray, same shape
    """
    processed = np.zeros_like(signals)
    for i in range(signals.shape[1]):
        lead = signals[:, i]
        # 1. Baseline wander removal (high-pass filter ~0.5 Hz)
        lead = bandpass_filter(lead, fs=fs, lowcut=0.5, highcut=40)
        # 2. Normalize lead
        lead = (lead - np.mean(lead)) / np.std(lead)
        processed[:, i] = lead
    return processed
```

```python
import numpy as np
from scipy.signal import find_peaks
from tqdm import tqdm

# Store features
rr_features = []

for idx, ecg in enumerate(tqdm(all_ecgs, desc="Extracting RR interval features")):
    rr_feats_per_ecg = []
    for lead in range(ecg.shape[1]):  # all 12 leads
        signal = ecg[:, lead]
        # Detect R-peaks (distance depends on sampling rate, fs=100, so ~0.6s -> 60 samples)
        peaks, _ = find_peaks(signal, distance=60)
        rr_intervals = np.diff(peaks)  # in samples
        # Convert to seconds
        rr_intervals_sec = rr_intervals / 100.0
        # Basic stats
        if len(rr_intervals_sec) > 0:
            rr_mean = np.mean(rr_intervals_sec)
            rr_std = np.std(rr_intervals_sec)
            rr_min = np.min(rr_intervals_sec)
            rr_max = np.max(rr_intervals_sec)
        else:
            rr_mean = rr_std = rr_min = rr_max = 0
        rr_feats_per_ecg.extend([rr_mean, rr_std, rr_min, rr_max])
    rr_features.append(rr_feats_per_ecg)

# Convert to array
X_rr = np.array(rr_features)
y = np.array(all_labels)

print("RR features shape:", X_rr.shape)
```

```python
import numpy as np
from scipy.signal import find_peaks
from tqdm import tqdm

# Store features
wave_features = []

for idx, ecg in enumerate(tqdm(all_ecgs, desc="Extracting P/QRS/T features")):
    feats_per_ecg = []
    
    for lead in range(ecg.shape[1]):  # all 12 leads
        signal = ecg[:, lead]
        
        # 1. QRS detection (R-peaks)
        r_peaks, _ = find_peaks(signal, distance=60)
        
        # Initialize arrays
        p_amp_list, qrs_amp_list, t_amp_list = [], [], []
        p_dur_list, qrs_dur_list, t_dur_list = [], [], []
        
        for i in range(1, len(r_peaks)-1):
            # QRS: around R-peak (±20 samples)
            qrs_window = signal[r_peaks[i]-20:r_peaks[i]+20]
            qrs_amp = np.max(qrs_window) - np.min(qrs_window)
            qrs_amp_list.append(qrs_amp)
            qrs_dur_list.append(len(qrs_window)/100.0)  # duration in sec
            
            # P-wave: before R-peak (10-20 samples)
            p_window = signal[r_peaks[i]-40:r_peaks[i]-10]
            if len(p_window) > 0:
                p_amp = np.max(p_window) - np.min(p_window)
                p_amp_list.append(p_amp)
                p_dur_list.append(len(p_window)/100.0)
            
            # T-wave: after R-peak (10-40 samples)
            t_window = signal[r_peaks[i]+10:r_peaks[i]+40]
            if len(t_window) > 0:
                t_amp = np.max(t_window) - np.min(t_window)
                t_amp_list.append(t_amp)
                t_dur_list.append(len(t_window)/100.0)
        
        # Aggregate statistics (mean, std)
        for arr in [p_amp_list, qrs_amp_list, t_amp_list, p_dur_list, qrs_dur_list, t_dur_list]:
            if len(arr) > 0:
                feats_per_ecg.append(np.mean(arr))
                feats_per_ecg.append(np.std(arr))
            else:
                feats_per_ecg.append(0)
                feats_per_ecg.append(0)
    
    wave_features.append(feats_per_ecg)

# Convert to array
X_wave = np.array(wave_features)
print("P/QRS/T features shape:", X_wave.shape)

#For each lead:
#QRS amplitude = peak-to-peak around R-peak
#P-wave amplitude = peak-to-peak before R-peak
#T-wave amplitude = peak-to-peak after R-peak
#Compute mean & std across beats → 6 values per waveform per lead

#12 leads × 6 features × 2 stats = 144 features per ECG
```

```python
import numpy as np
from scipy.signal import find_peaks
from tqdm import tqdm

# Store HRV features for all ECGs
hrv_features = []

for idx, ecg in enumerate(tqdm(all_ecgs, desc="Extracting HRV features")):
    # Use Lead II (commonly used for HRV)
    signal = ecg[:, 1]
    
    # 1. Detect R-peaks
    r_peaks, _ = find_peaks(signal, distance=60)  # distance ~ 60 samples (0.6s at 100Hz)
    
    # 2. Compute RR intervals (in seconds)
    rr_intervals = np.diff(r_peaks) / 100.0  # sampling freq = 100 Hz
    
    if len(rr_intervals) == 0:  # handle empty case
        hrv_feats = [0]*4
    else:
        # 3. HRV metrics
        mean_rr = np.mean(rr_intervals)              # Mean RR interval
        sdnn = np.std(rr_intervals)                  # Standard deviation of RR intervals
        rmssd = np.sqrt(np.mean(np.diff(rr_intervals)**2))  # Root mean square of successive differences
        pnn50 = np.sum(np.abs(np.diff(rr_intervals)) > 0.05) / len(rr_intervals)  # proportion >50ms
        
        hrv_feats = [mean_rr, sdnn, rmssd, pnn50]
    
    hrv_features.append(hrv_feats)

# Convert to array
X_hrv = np.array(hrv_features)
print("HRV features shape:", X_hrv.shape)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
from tqdm import tqdm

# List to store features for all ECGs
stat_features = []

for idx, ecg in enumerate(tqdm(all_ecgs, desc="Extracting statistical features")):
    features = []
    # Loop over each lead (12 leads)
    for lead_idx in range(ecg.shape[1]):
        lead = ecg[:, lead_idx]
        # Basic statistical features
        features.append(np.mean(lead))           # mean
        features.append(np.std(lead))            # std
        features.append(np.min(lead))            # min
        features.append(np.max(lead))            # max
        features.append(skew(lead))              # skewness
        features.append(kurtosis(lead))          # kurtosis
    stat_features.append(features)

# Convert to DataFrame for easy handling
stat_features_df = pd.DataFrame(stat_features, 
                                columns=[f"L{lead+1}_{feat}" 
                                         for lead in range(12) 
                                         for feat in ["mean", "std", "min", "max", "skew", "kurt"]])
print("Statistical features shape:", stat_features_df.shape)

#Resulting shape: 21388 x 72 → 12 leads × 6 features each.
#Why important: Captures signal distribution, extreme values, and shape characteristics, which can help classifiers detect abnormalities.
```

```python
#📊 Frequency-Domain HRV Features
#We’ll compute power spectral density (PSD) of RR intervals using Welch’s method:
#-->VLF power: < 0.04 Hz (ignored sometimes in short-term ECGs, but we’ll include if data length permits).
#-->LF power: 0.04–0.15 Hz → sympathetic + parasympathetic balance.
#-->HF power: 0.15–0.4 Hz → parasympathetic (respiratory sinus arrhythmia).
#-->LF/HF ratio: balance between sympathetic & parasympathetic systems.
#These are standard HRV features used in clinical studies.

#✅ Code to Extract Frequency HRV Features

import numpy as np
from scipy.signal import welch, find_peaks
from tqdm import tqdm

# Store frequency-domain HRV features
hrv_freq_features = []

fs_rr = 4.0  # resample RR time series at 4 Hz

for idx, ecg in enumerate(tqdm(all_ecgs, desc="Extracting HRV frequency features")):
    # Use Lead II for consistency
    signal = ecg[:, 1]
    
    # Step 1: Detect R-peaks
    r_peaks, _ = find_peaks(signal, distance=60)  # ~0.6s at 100 Hz
    
    # Step 2: RR intervals (seconds)
    rr_intervals = np.diff(r_peaks) / 100.0  # fs = 100 Hz
    
    if len(rr_intervals) < 4:
        hrv_freq_features.append([0,0,0,0])
        continue
    
    # Step 3: Interpolate RR intervals into evenly spaced signal
    rr_times = np.cumsum(rr_intervals)  # cumulative time
    rr_series = rr_intervals
    
    uniform_time = np.arange(0, rr_times[-1], 1/fs_rr)
    rr_interp = np.interp(uniform_time, rr_times, rr_series)  # FIXED
    
    # Step 4: PSD (Welch)
    freqs, psd = welch(rr_interp, fs=fs_rr, nperseg=min(256, len(rr_interp)))
    
    # Step 5: Integrate PSD in frequency bands
    vlf_band = (freqs >= 0.003) & (freqs < 0.04)
    lf_band  = (freqs >= 0.04) & (freqs < 0.15)
    hf_band  = (freqs >= 0.15) & (freqs < 0.40)
    
    vlf_power = np.trapz(psd[vlf_band], freqs[vlf_band]) if np.any(vlf_band) else 0
    lf_power  = np.trapz(psd[lf_band], freqs[lf_band]) if np.any(lf_band) else 0
    hf_power  = np.trapz(psd[hf_band], freqs[hf_band]) if np.any(hf_band) else 0
    
    # Step 6: LF/HF ratio
    lf_hf_ratio = lf_power / hf_power if hf_power > 0 else 0
    
    hrv_freq_features.append([vlf_power, lf_power, hf_power, lf_hf_ratio])

# Convert to numpy array
X_hrv_freq = np.array(hrv_freq_features)

print("HRV frequency features shape:", X_hrv_freq.shape)  
```
