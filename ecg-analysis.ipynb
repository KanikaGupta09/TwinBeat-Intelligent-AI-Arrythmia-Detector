```python
sns.kdeplot(data['V5'], fill=False, color='r', alpha=0.4)

plt.show()
```

```python
sns.pairplot(data)

plt.show()
```

```python
plt.figure(figsize=(15, 4))
plt.plot(data['time_ms'], data['MLII'], color='blue')
plt.title("ECG Signal - MLII Lead")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
data.head()
```

```python
# Subplut
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.hist(data['time_ms'], bins=30)
plt.title('time_ms')
plt.xlabel('time_ms')
plt.ylabel('Frequency')

plt.subplot(3, 1, 2)
sns.countplot(data['MLII'])
plt.title('MLII') 
plt.xlabel('MLII')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
sns.kdeplot(data['V5'], color='red', fill=True)
plt.title('V5')
plt.xlabel('V5')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
